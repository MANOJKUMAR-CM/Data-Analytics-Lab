```python
import numpy as np
import matplotlib.pyplot as plt
```

Unipolar sigmoid function =
    f(x): $\frac{1}{1+e^{-x}}$

```python
def unipolar_sigmoid(x):
    return 1/ (1+np.exp(-x))
```

Bipolar sigmoid function = 
f(x): $\frac{1-e^{-x}}{1+e^{-x}}$ = $\frac{2}{1+e^{-x}} - 1$


```python
def bipolar_sigmoid(x):
    return 2*unipolar_sigmoid(x) -1
```

tanh = f(x): $\frac{e^{x} - e^{-x}}{e^{x} + e^{-x}} $

```python
def tanh(x):
    return np.tanh(x)
```

```python
a = [-5, -1, -.1, -.01, .001, .01, .1, 1, 5]
x = np.linspace(-10,10,100)
```

```python
plt.figure(figsize=(10,10))
for i in a:
    plt.plot(x,bipolar_sigmoid(i*x),'--',label=f"Bipolar_Sigmoid: a={i}")
    plt.plot(x,tanh(i*x),label=f"tanh: a={i}")
plt.title('Comparison of bipolar_sigmoid(ax) and tanh(ax) for different values of a')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('Function {bipolar_sigmoid and tanh} output')
plt.legend()
plt.show()
```

```python
# Linear range of X: The range of values of x for which the function behaves linear
```

```python
a = [-5, -1, -.1, -.01, .001, .01, .1, 1, 5]
x = np.linspace(-10,10,1000)

def linear_approx(i,x):
    return i*x
```

```python
X_linear = []
for i in a:
    y_bipolar_sigmoid = bipolar_sigmoid(i*x)
    y_linear = linear_approx(i,x)
    diff = np.abs(y_bipolar_sigmoid - y_linear)
    index = np.where(diff<0.1)[0] # Looking for points within 10% deviation
    linear_range = round(x[index[0]],2), round(x[index[-1]],2)
    X_linear.append(linear_range)
    
```

```python
for i in range(len(a)):
    print(f"Linear range of X for a= {a[i]}: ",X_linear[i])
    print()
```
